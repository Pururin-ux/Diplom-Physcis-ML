# 00 Single Dot Sanity

Minimal manual sanity checks for the rectangular quantum dot solver.

This notebook:
- runs one example dot calculation,
- prints the first 4 energies,
- performs a tiny sweep over `Lx` at fixed `Ly`,
- plots how the first 4 levels change.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import numpy as np

PROJECT_ROOT = Path.cwd().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.geometry import build_rectangular_dot
from src.kwant_solver import lowest_four_energies
```

```python
# One example rectangular dot calculation.
Lx, Ly = 12, 10
fsys = build_rectangular_dot(Lx=Lx, Ly=Ly)
energies = lowest_four_energies(Lx=Lx, Ly=Ly)

print(f"Configuration: Lx={Lx}, Ly={Ly}")
print(f"Number of sites: {len(fsys.sites)}")
print("First 4 energies:")
for i, e in enumerate(energies):
    print(f"E{i} = {e:.10f}")
```

```python
# Tiny sweep over Lx at fixed Ly.
Ly_fixed = 10
lx_values = np.arange(8, 15)

levels = np.array([lowest_four_energies(Lx=int(lx), Ly=Ly_fixed) for lx in lx_values])

plt.figure(figsize=(7, 4.5))
for level_idx in range(4):
    plt.plot(lx_values, levels[:, level_idx], marker="o", label=f"E{level_idx}")

plt.xlabel("Lx (lattice units)")
plt.ylabel("Energy (model units)")
plt.title(f"First 4 levels vs Lx at Ly={Ly_fixed}")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
